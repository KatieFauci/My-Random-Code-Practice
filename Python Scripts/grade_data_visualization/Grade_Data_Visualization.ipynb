```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_grades(grades_file):
    df = pd.read_excel(grades_file)

    # Calculate Datapoints
    avg = df['Grade'].mean()
    max = df['Grade'].max()
    min = df['Grade'].min()

    print(f'Average Grade: {avg}')
    print(f'Highest Grade: {max}')
    print(f'Lowest Grade: {min}')

    # Create Bell Plot
    stdev = df['Grade'].std()

    plt.figure(figsize=(10,6))
    sns.kdeplot(df['Grade'], color='blue', fill=True)

    plt.title('Bell Curve of Grades')
    plt.xlabel('Grades')
    plt.ylabel('Density')

    plt.grid()
    plt.show()


visualize_grades("Grades_1.xlsx")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_grades(grades_file):
    df = pd.read_excel(grades_file)

    # Calculate Datapoints
    avg = df['Grade'].mean()
    max = df['Grade'].max()
    min = df['Grade'].min()

    print(f'Average Grade: {avg}')
    print(f'Highest Grade: {max}')
    print(f'Lowest Grade: {min}')

    sorted_df = df.sort_values(by='Grade', ascending=False)
    top3 = sorted_df.head(3)
    bottom3 = sorted_df.tail(3)

    print('\nTop 3 Students:')
    print(top3.to_string(index=False))    

    print('\nBottom 3 Students:')
    print(bottom3.to_string(index=False))    

    # Create Bell Plot
    stdev = df['Grade'].std()

    plt.figure(figsize=(10,6))
    sns.kdeplot(df['Grade'], color='blue', fill=True)

    plt.title('Bell Curve of Grades')
    plt.xlabel('Grades')
    plt.ylabel('Density')

    plt.grid()
    plt.show()


visualize_grades("Grades_2.xlsx")
```
